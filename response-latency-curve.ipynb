## Now parse Manual and Auto Rec Data

### Manual

```python
##IMPORTS#######
import os 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
################

ALL_SUBS_MAN = []
uniqueID_man = []

ALL_ONSETS_MAN = []

d = '../data/PennTotalRecall/'

for subject in os.listdir(d):
    if not subject.startswith('.'):
        uniqueID_man.append(subject)
        SUB_LISTS = []
        SUB_ONSETS = []
        
        #change into subject directory
        path = d+subject
        os.chdir(path)
        files = os.listdir(os.curdir)
        #print files
        files = files[0:]
        for fl in files:
            if not fl.startswith('.'):
                #print fl
                MAN = []

                with open(fl) as f:
                    MAN_contents = f.read().split()
                for line in MAN_contents[0:]:
                    MAN.append(line.upper())

                words = MAN[2::3]
                onsets = MAN[0::3]

                SUB_LISTS.append(words)
                SUB_ONSETS.append(onsets)
        ALL_SUBS_MAN.append(SUB_LISTS)
        ALL_ONSETS_MAN.append(SUB_ONSETS)

        os.chdir('../../')
```

## Auto

```python
##IMPORTS#######
import os 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
################

ALL_SUBS_AUTO = []
uniqueID_auto = []

ALL_ONSETS_AUTO = []

d = '../data/Google/'

for subject in os.listdir(d):
    if not subject.startswith('.'):
        uniqueID_auto.append(subject)
        SUB_LISTS = []
        SUB_ONSETS = []
        
        #change into subject directory
        path = d+subject
        os.chdir(path)
        files = os.listdir(os.curdir)
        files = files[0:]
        for fl in files:
            if not fl.startswith('.'):
                AUTO = []

                with open(fl) as f:
                    AUTO_contents = f.readlines()
                    #print(AUTO_contents)
                words = []
                onsets = []
                for line in AUTO_contents[0:]:
                    #print line
                    word = line.split(',')[0]
                    onset = line.split(',')[1]
                    words.append(word)
                    AUTO.append(line.upper())
                    onsets.append(onset)
#                 words = AUTO[0]
                #onsets = AUTO[1]

                SUB_LISTS.append(words)
                SUB_ONSETS.append(onsets)
        ALL_SUBS_AUTO.append(SUB_LISTS)
        ALL_ONSETS_AUTO.append(SUB_ONSETS)

        os.chdir('../../')
```

```python
# make this into df that matches blandEgg.pres subject order

for x,subject in enumerate(blandEgg.meta['ids']):
    
    sub_data = ALL_SUBS_AUTO[uniqueID_auto.index(str(blandEgg.meta['ids'][x]))]
    sub_index = x
    
    if x==0:
        df = pd.DataFrame(sub_data)
    elif x==1 :
        df = pd.concat([df, pd.DataFrame(sub_data)], keys=[0,x])
    else:
        df = pd.concat([df, pd.concat([pd.DataFrame(sub_data)], keys=[x])])
        
for x,subject in enumerate(room1Egg.meta['ids']):
    sub_data = ALL_SUBS_AUTO[uniqueID_auto.index(str(room1Egg.meta['ids'][x]))]
    sub_index = x + 28
    
    df = pd.concat([df, pd.concat([pd.DataFrame(sub_data)], keys=[sub_index])])
```
