```python
funkyPost_list = []

for index, row in df_post.iterrows():
    
    sub = []
    j = [row["Timestamp"], row['Subject ID'],] 
    

    if type(row['Subject ID']) == float:
        j.append("subID == nan")
    
    else: 
        if row['Subject ID'][0:4] != 'FRFR':
            j.append("subID[0:4] != FRFR")
                 
    if type(row['Experimenter Name']) == float:
        j.append("expName = nan")
    
#     if row['How tired do you feel?'] == 'Very tired':
#         j.append("Very Tired")
    
    if type(row['How tired do you feel?']) == float:
        j.append("Tired: no answer")
        
    if type(row['What strategy did you use to remember the lists of words?']) == float:
        j.append('Strategy: no answer')
     
    if type(row["How engaging did you find this experiment?"]) == float:
        j.append('Engaging: no answer')
        
    
    if len(j) > 2:
        funkyPost_list.append(j)
                
    
funkyPost = pd.DataFrame(funkyPost_list)
    
```
