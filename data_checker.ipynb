```python
funkyPost_list = []

for index, row in df_post.iterrows():
    
    sub = []
    j = [row["Timestamp"], row['Subject ID'],] 
    

    if type(row['Subject ID']) == float:
        j.append("subID == nan")
    
    else: 
        if row['Subject ID'][0:4] != 'FRFR':
            j.append("subID[0:4] != FRFR")
                 
    if type(row['Experimenter Name']) == float:
        j.append("expName = nan")
    
    if row['How tired do you feel?'] == 'Very tired':
        j.append("Very Tired")
    
    if type(row['How tired do you feel?']) == float:
        j.append("Tired: no answer")
        
    if type(row['What strategy did you use to remember the lists of words?']) == float:
        j.append('Strategy: no answer')
     
    if type(row["How engaging did you find this experiment?"]) == float:
        j.append('Engaging: no answer')
        
    
    if len(j) > 2:
        funkyPost_list.append(j)
                
    
funkyPost = pd.DataFrame(funkyPost_list)
    
```

```python
funkyPre_list = []

for index, row in df_pre.iterrows():
    
    sub = []
    j = [row["Timestamp"], row['Subject ID'],] 
    


    if type(row['Experimenter Name']) == float:
        j.append("expName = nan")

    if type(row['Age:']) == float:
        j.append("no age")

    if type(row['Sex']) == float: 
        j.append("no sex")

    if type(row['Ethnicity']) == float:
        j.append("no eth")

    if type(row['Race (check all that apply)']) == float:
        j.append("no race")

    if row['Do you have any reading impairments (e.g. dyslexia, uncorrected far-sightedness, etc.)? (yes/no)'] != "No":
        if type(row['Do you have any reading impairments (e.g. dyslexia, uncorrected far-sightedness, etc.)? (yes/no)']) == float:
            j.append("no read impair info")
        else:
            j.append("read impaired")

    if row["Are you taking any medications or have you had any recent injuries that could affect your memory or attention? (if so, describe below)"] != "No":         
        if row["Are you taking any medications or have you had any recent injuries that could affect your memory or attention? (if so, describe below)"] == "Yes":
            j.append('meds'+str(row['If yes above, describe']))

        else:
            j.append('no med info')


    if len(j) > 2:
        funkyPre_list.append(j)
                
    
funkyPre = pd.DataFrame(funkyPre_list)
```
