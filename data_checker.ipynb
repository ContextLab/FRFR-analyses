<h2> Analyze Post Questionnaire</h2>

```python
for index, row in df_post.iterrows():
    
    if row['Subject ID'] == float:
        #print("missing subject ID: ")
        print("missing subject ID: ", row["Timestamp"], row["subject ID"])
        print
    
    elif row['Subject ID'] == str:
        if row['Subject ID'][0:4] != 'FRFR':
            #print("Subject ID not starting with FRFR: ")
            print("Subject ID not starting with FRFR: ", row["Timestamp"], row["Subject ID"])
            print
    else:
        print(row["Timestamp"], row["Subject ID"])
        
    if row['Experimenter Name'] == float:
        #print("No experimenter name: ")
        print("No experimenter name: ", row["Timestamp"], row["Subject ID"])
        print
        
    if row['How tired do you feel?'] == 'Very tired':
        #print("Very tired: ")
        print("Very tired: ", row["Timestamp"], row["Subject ID"])
        print
    
    if row['How tired do you feel?'] == float:
        #print("Did not answer how tired: ")
        print("Did not answer how tired: ", row["Timestamp"], row["Subject ID"])
        print
        
    if row['What strategy did you use to remember the lists of words?'] == float:
        #print("Did not answer strategy question: ")
        print("Did not answer strategy question: ", row["Timestamp"], row["Subject ID"])
        print
    
```
